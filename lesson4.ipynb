Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.keys()
```

```python
from sklearn.model_selection import train_test_split
```

```python
data = boston.data

target = boston.target

feature_names = boston.feature_names
```

```python
X = pd.DataFrame(data, columns=feature_names)

y = pd.DataFrame(target, columns=['price'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
lr_score= r2_score(y_test, y_pred)
lr_score
```

Создайте список high_corr из признаков, корреляция которых с полем target по абсолютному 
значению превышает 0.5 
(причем, само поле target не должно входить в этот список).

```python
h_t= X_corr.loc[(X_corr.index!='target') ,'target']

high_corr =h_t[np.fabs(h_t)>0.5].index
high_corr
```

```python
X.drop('target',axis=1, inplace=True)
```

```python
for f in high_corr:
    X[f+'_2']= X[f]**2
```
